# Missing Data Handling and Feature Engineering

## Split Train Test Data

In order to prevent data leakage for the machine learning model, we split the data prior to imputing missing data.

```python
#Target and Feature selection 
X = df.drop('RainTomorrow', axis=1)
Y = df['RainTomorrow']
```

```python
#Split train test 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=df['RainTomorrow'], random_state=42)
```

```python
#Merge Train data and Test data 
df_train = pd.concat([X_train, y_train], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)
df_train.shape, df_test.shape
```

## Imputing Missing Data for Numerical Feature

```python
#Train missing data percentage
(df_train.isnull().sum()/len(df_train))*100
```

```python
#Test mssing data percentage
(df_test.isnull().sum()/len(df_test))*100
```

```python
#Numerical Column Missing Data Imputing (Using Median)
for i in numerical: 
    df_train[i].fillna(df_train[i].median(), inplace=True)
    df_test[i].fillna(df_train[i].median(), inplace=True)
```

```python
#Recheck Missing value both in training and Test Data Set 
print('Percentage of Training Data Missing Value:')
print((df_train.isnull().sum()/len(df_train))*100)
print("")
print('Percentage of Testing Data Missing Value:')
print((df_test.isnull().sum()/len(df_test))*100)
```

## Imputing Missing Data for Categorical Feature

```python
#Impute Categorical Missing Data With mode
for i in categorical:
    df_train[i].fillna(df_train[i].mode()[0], inplace=True)
    df_test[i].fillna(df_train[i].mode()[0], inplace=True)
```

```python
#Recheck Missing value both in training and Test Data Set 
print('Percentage of Training Data Missing Value:')
print((df_train.isnull().sum()/len(df_train))*100)
print("")
print('Percentage of Testing Data Missing Value:')
print((df_test.isnull().sum()/len(df_test))*100)
```

## Converting Categorical to Numerical Variable

```python
#Count for categorical feature unique value
for i in categorical: 
    print(f'{i} Number of Unique Variable: {df[i].nunique()}')
    print("")
```

### Encoding for RainToday and RainTommorrow Feature

Since there are only two unique value in this feature, we will use label encoding for transforming the feature to numerical (Yes equal to one, and No equal to zero)

```python
#For RainToday and RainTommorrow Feature 
enc_y_n = {'Yes': 1, 'No' : 0}
df_train['RainTomorrow'] = df_train['RainTomorrow'].map(enc_y_n)
df_test['RainTomorrow'] = df_test['RainTomorrow'].map(enc_y_n)
df_train['RainToday'] = df_train['RainToday'].map(enc_y_n)
df_test['RainToday'] = df_test['RainToday'].map(enc_y_n)
```

```python
df_train['RainTomorrow'], df_test['RainTomorrow']
```

```python
df_train['RainToday'], df_test['RainToday']
```

### Encoding for Wind Direction Feature

Since there is 16 unique value for this feature, we will apply mean target encoding to transform the data to numerical. 

#### WindGustDir Feature

```python
#Creating Total Occurents In table 
freq_wgd = df_train['WindGustDir'].value_counts().reset_index()
freq_wgd.rename(columns= {'index':'WindGustDir', 'WindGustDir' : 'Frequency'}, inplace=True)
freq_wgd
```

```python
#Creating Total Occurent Based on target 
sum_wgd = df_train.groupby('WindGustDir')['RainTomorrow'].sum().reset_index()
sum_wgd
```

```python
#Merging data 
enc_windgust = pd.merge(freq_wgd,sum_wgd)
#Create Mean enc data columns
enc_windgust['mean_enc_windgust'] = round(enc_windgust['RainTomorrow']/enc_windgust['Frequency'],2)
enc_windgust = enc_windgust[['WindGustDir', 'mean_enc_windgust']]
enc_windgust
```

```python
#Merge with train data and test data 
df_train = pd.merge(df_train,enc_windgust)
df_test = pd.merge(df_test,enc_windgust)
```

```python
#Inspect Train Data 
df_train.head()
```

```python
#Inspect test data 
df_test.head()
```

# Feature Selection

For feature selection, we will use embedded method. This will be done by deploying Logistic Regression Model and see which feature has the biggest coefficient (Bigger than the Mean Coefficient).

## Scalling Numerical Data and Split Feature and Target

```python
#Split X and Y 
X_train = df_train.drop('RainTomorrow', axis=1)
y_train = df_train['RainTomorrow']
X_test = df_test.drop('RainTomorrow', axis=1)
y_test = df_test['RainTomorrow']
```

```python
#Split Numerical and Categorical on Feature 
X_train_num = X_train[numerical]
X_train_cat = X_train.drop(numerical, axis=1)
X_test_num = X_test[numerical]
X_test_cat = X_test.drop(numerical, axis=1)
```

```python
#Scalling Numerical Data 
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
#fit to train data
ss.fit(X_train_num)
X_train_num_scld = pd.DataFrame(ss.transform(X_train_num),columns = X_train_num.columns)
X_test_num_scld = pd.DataFrame(ss.transform(X_test_num),columns = X_test_num.columns)
```

```python
#Merge Back Data Set 
X_train_scld = pd.concat([X_train_num_scld, X_train_cat], axis=1)
X_test_scld = pd.concat([X_test_num_scld, X_test_cat], axis=1)
```

```python
#Inspect Train Data Set 
X_train_scld
```

```python
#Inspect Test Data Set 
X_test_scld
```

## Feature Selection Using Embedded Method

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

sel_ = SelectFromModel(
    LogisticRegression(C=1000, penalty='l2', max_iter=300, random_state=42))

sel_.fit(X_train_scld, y_train)
```

```python
#Selected Features 
selected_feat = X_train_scld.columns[(sel_.get_support())]
selected_feat
```

```python
# Since the pressure at 9 am is Highly Correlated with Pressure at 3 PM Thus we will evaluate which feature is going to be droped.
for_cor = pd.concat([X_train_scld, y_train], axis=1)
for_cor.corr()
```

Conclusion: Since Pressure at 9 am having a greater correlation with the target, We are going to drop the Pressure at 3 pm data.

```python
#Apply Selected Feature on Training and Test Data Set 
selected_feature = ['WindGustSpeed', 'Humidity3pm', 'Pressure9am', 'RainToday', 'mean_enc_windgust']
X_train_model = X_train_scld[selected_feature]
X_test_model = X_test_scld[selected_feature]
```

# Model Building (Baseline Model)

## Evaluate Best Model (Logistic Regression or SVM)

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, roc_auc_score

model_alg = [SVC(), LogisticRegression()]
df_mod_eval = pd.DataFrame(columns = ['Algorithm','AUC Score'])
for model in model_alg : 
    model.fit(X_train_model, y_train)
    algorithm = str(type(model)).split('.')[-1][:-2]
    y_pred = model.predict(X_test_model)
    auc = roc_auc_score(y_test, y_pred)
    df_mod_eval = df_mod_eval.append({'Algorithm' : algorithm, 
                                    'AUC Score' : auc},ignore_index=True)
df_mod_eval
```

Conclusion : LogisticRegression performed better than Support Vector Machine. Thus, we will choose LogisticRegression for further development.

## Best Line Model Evaluation

```python
#Confusion Matrix 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, model_alg[1].predict(X_test_model))
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.title('Confusion Matrix', fontweight = 'bold')
```

```python
#Classification Report 
from sklearn.metrics import classification_report
print(classification_report(y_test, model_alg[1].predict(X_test_model)))
```

Conclusion: The model has sufficient accuracy. But it could be improved.

# Model Improvements

## Imbalanced Data Handling

```python
#Perform SMOTE Imbalanced Data Handling 
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train_blncd, y_train_blncd = smote.fit_resample(X_train_model, y_train)
```

```python
X_train_blncd.shape , y_train_blncd.shape
```

## Train Model using Balanced Data

```python
#Train model
model_alg[1].fit(X_train_blncd, y_train_blncd)
```

## Improved Model Evaluation

```python
#Confusion Matrix 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, model_alg[1].predict(X_test_model))
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.title('Confusion Matrix', fontweight = 'bold')
```

```python
#Classification Report 
from sklearn.metrics import classification_report
print(classification_report(y_test, model_alg[1].predict(X_test_model)))
```

```python
#ROC AUC Curve Plot 
y_pred_proba = model_alg[1].predict_proba(X_test_model)[:][:,1]

df_actual_predicted = pd.concat([pd.DataFrame(np.array(y_test), columns=['y_actual']), pd.DataFrame(y_pred_proba, columns=['y_pred_proba'])], axis=1)
df_actual_predicted.index = y_test.index
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr, tpr, tr = roc_curve(df_actual_predicted['y_actual'], df_actual_predicted['y_pred_proba'])
auc = roc_auc_score(df_actual_predicted['y_actual'], df_actual_predicted['y_pred_proba'])

plt.plot(fpr, tpr, label='AUC = %0.4f' %auc)
plt.plot(fpr, fpr, linestyle = '--', color='k')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
```

## Cross Validation Evaluation

```python
#Cross Validation Score Evaluation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model_alg[1], X_train_blncd, y_train_blncd, cv=10, scoring='roc_auc')
print(scores)
```

Conclusion : The difference between cross validation score is not that big thus we can conclude that it is quite enough.
